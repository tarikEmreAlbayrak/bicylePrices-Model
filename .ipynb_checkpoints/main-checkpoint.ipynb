## veriyi test/train olarak split etmek

```python
from sklearn.model_selection import train_test_split
```

```python
data["Fiyat"].values # np dizisine çeviriyor
```

```python
# y = wx + b
# y -> label
y = data["Fiyat"].values

# x -> feature (özellik)
x = data[["BisikletOzellik1", "BisikletOzellik2"]].values

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape # 670 tane veri trainin içine geldi
```

```python
x_test.shape # yüzde 33 ü teste kalanı traine
```

## 2. aşama normalize

```python
#scaling, boytunu değişmek küçültmek vs
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(4, activation="relu") )
model.add(Dense(4, activation="relu"))
model.add(Dense(4, activation="relu") )
# 4 nörondan 3 tane gizli katman
model.add(Dense(1)) # 1 tane output katmanı
model.compile(optimizer="rmsprop", loss="mse") # mse : mean squared error
# loss : cost function (adam en iyi performans veren)
          
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss = model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)), y = loss)
```

```python
plt.plot(loss)
```

```python
trainLoss = model.evaluate(x_train,y_train, verbose=0)
```

```python
testLoss = model.evaluate(x_test,y_test, verbose=0)
```

```python
trainLoss
```

```python
testLoss
```
